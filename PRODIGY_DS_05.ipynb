```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Road.csv')
df
```

```python
# Check for missing values
print("Missing values in the dataset:")
print(df.isnull().sum())
```

```python
# Data Cleaning
# Handling missing values
df.dropna(inplace=True)  # Remove rows with missing values
print(df.isnull().sum())
```

```python
# Check data types and basic statistics
print("Data types and basic statistics:")
print(df.info())
```

```python
print(df.describe())
```

```python
# Visualizing accidents by day of the week
plt.figure(figsize=(10, 6))
sns.countplot(x='Day_of_week', data=df)
plt.title('Accidents by Day of the Week')
plt.show()
```

<h6><b>Inference :</b> The above graph shows that the most number of accidents happened on Friday.</h6>

```python
# Visualizing accidents by gender of the driver
plt.figure(figsize=(10, 6))
sns.countplot(x='Sex_of_driver', data=df)
plt.title('Accidents by Gender of the Driver')
plt.show()
```

<h6><b>Inference :</b> The above graph shows that more number Males are involved in the accident.</h6>

```python
# Visualizing accidents by driving experience of the driver
plt.figure(figsize=(12, 6))
sns.countplot(x='Driving_experience', data=df)
plt.title('Accidents by Driving Experience of the Driver')
plt.show()
```

<h6><b>Inference :</b> The above graph shows that the drivers that are involved in the accidents have a driving experience between 5 to 10 years.</h6>

```python
# Visualizing accidents by time of day
plt.figure(figsize=(20, 8))
ax = sns.countplot(x='Time', data=df, order=sorted(df['Time'].unique()))
ax.set_xticks(ax.get_xticks()[::10])  # Show every 10th tick
plt.title('Accidents by Time of Day')
plt.xticks(rotation=90)
plt.show()
```

```python
# Visualizing accidents by time of day
plt.figure(figsize=(20, 8))
ax = sns.countplot(x='Time', data=df, order=sorted(df['Time'].unique()))
ax.set_xticks(ax.get_xticks()[::10])  # Show every 10th tick
plt.title('Accidents by Time of Day')
plt.xticks(rotation=90)
plt.show()
```

<h6><b>Inference :</b> The above graph shows that most of the accidents happen either during noon or early evening.</h6>

```python
# Select only numeric columns
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns

# Compute the correlation matrix for numeric columns
correlation_matrix = df[numeric_columns].corr()

# Display the correlation matrix
print(correlation_matrix)
```

```python
# Select only numeric columns
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns

# Compute the correlation matrix for numeric columns
correlation_matrix = df[numeric_columns].corr()

# Plot the correlation heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.show()
```

```python
# Calculate the average severity by weather condition
average_severity_by_weather = df.groupby('Weather_conditions')['Accident_severity'].value_counts(normalize=True).unstack().mean(axis=1)

# Display the result
print("Average Severity by Weather Condition:")
print(average_severity_by_weather)
```

<h6><b>Inference :</b> The above results show that the average severity of the accident is the maximum whrn the weather conditions are 'Raining and Windy'
</h6>
